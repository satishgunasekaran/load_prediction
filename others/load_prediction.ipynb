```python
# CatBoost parameters
ctb_reg_params = {
    'learning_rate':   
      hp.choice('learning_rate', np.arange(0.05, 0.31, 0.05)),
    'max_depth':        
     hp.choice('max_depth', np.arange(5, 16, 1, dtype=int)),
    'colsample_bylevel': 
    hp.choice('colsample_bylevel', np.arange(0.3, 0.8, 0.1)),
    'n_estimators': 100,
    'eval_metric': 'RMSE',
}
ctb_fit_params = {
    'early_stopping_rounds': 10,
    'verbose': False
}
ctb_para = dict()
ctb_para['reg_params'] = ctb_reg_params
ctb_para['fit_params'] = ctb_fit_params
ctb_para['loss_func' ] = lambda y, pred: np.sqrt(mean_squared_error(y, pred))
```

```python
class CatHPOptimizer(object):
    def __init__(self, x_train, x_test, y_train, y_test):
        self.x_train = x_train
        self.x_test = x_test
        self.y_train = y_train
        self.y_test = y_test
    def process(self, fn_name, space, trials, algo, max_evals):
        fn = getattr(self, fn_name)
        try:
            result = fmin(fn=fn, 
                          space=space, 
                          algo=algo,
                          max_evals=max_evals,
                          trials=trials)
        except Exception as e:
            return {'status': STATUS_FAIL,
                    'exception': str(e)}
        return result, trials
    
    def cat_reg(self, para):
        reg = ctb.CatBoostRegressor(**para['reg_params'])
        return self.train_reg(reg, para)
    
    def train_reg(self, reg, para):
        reg.fit(self.x_train, self.y_train,
                eval_set = [(self.x_train, self.y_train),
                            (self.x_test, self.y_test)],
                **para['fit_params'])
        pred = reg.predict(self.x_test)
        loss = para['loss_func'](self.y_test, pred)
        return {'loss':loss, 'status': STATUS_OK}
```

```python
train_x = train_features
test_x = test_features
train_y = train_labels
test_y = test_labels
```

```python
opti_model = CatHPOptimizer(train_x, test_x, train_y, test_y)
print(opti_model)
ctb_opt = opti_model.process(fn_name='cat_reg',
                             space=ctb_para, 
                             trials=Trials(),
                             algo=tpe.suggest, 
                             max_evals=100)
```

```python
best_params = {'learning_rate' :
               np.arange(0.05, 0.31, 0.05)[ctb_opt[0]['learning_rate']], 
               'max_depth' : 
               np.arange(5, 16, 1, dtype=int)[ctb_opt[0]['max_depth']],
               'colsample_bylevel':
               np.arange(0.3, 0.8, 0.1)[ctb_opt[0]['colsample_bylevel']]}
```

```python
CTB_model = ctb.CatBoostRegressor(verbose=0, n_estimators=100, 
                          colsample_bylevel=best_params['colsample_bylevel'],
                          learning_rate=best_params['learning_rate'],
                          max_depth=best_params['max_depth'],
                          early_stopping_rounds=10)
ctb_history= CTB_model.fit(train_features, train_labels)
```
